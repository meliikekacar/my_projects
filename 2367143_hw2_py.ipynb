```python
#          EXERCISE

def process_course_requests(course_catalog: dict, student_requests: list ):
  accepted= []
  rejected= []
  '''
we define a function named process_course_requests.
then we clarify that course_catalog is dictionary and student_requests contains list.
there is no course that is accepted or rejected so these are empty.
  '''
  for requested_course in student_requests:
    start_new, end_new = course_catalog[requested_course]
    conflict_found= False
    for existing_course in accepted:
        start_exist, end_exist = course_catalog[existing_course]

        if not (end_new <= start_exist or start_new >= end_exist):
          conflict_found= True
          rejected.append(f"{requested_course} rejected due to confliction with {existing_course}.")
          break

    if not conflict_found:
      accepted.append(requested_course)
  return accepted, rejected

'''
with for loop we check the courses that student required. if there is a conflict
then conflict_found turns true. the program look for accepted courses with starting
and ending times. we check requested courses' timelines with the already existed
timelines. For instance if a student requires a 9.00-11.00 class but they already
taken a 10.00-12.00 class. then program rejects that request. start_exist and end_exist
means the timelines of the course_catalog courses and start_new and end_new means the
timelines of the student_requests courses. then if conflict cannot be found the course
will be added to accepted courses list.
'''
course_catalog= {}
course_catalog_codes= input("What are the codes of the courses? ").split()
course_catalog_start_times= input("What are the start times of the courses? (e.g 900) ").split()
course_catalog_end_times= input("What are the end times of the courses? (e.g 1100) ").split()

'''
in the input part the program asks the student about all details like codes, starting
and ending times. with split() python turns shifts into lists.
'''


for i in range (len(course_catalog_codes)):
  code= course_catalog_codes[i]
  start= int(course_catalog_start_times[i])
  end= int(course_catalog_end_times[i])
  course_catalog[code] = (start, end)

'''
we must write start and ending times in integer for the program to work properly.
because otherwise python reads the input as strings and we cannot compare them.
but the main role of this block is that it combines all the data under course_catalog
dictionary.
'''

student_requests= input("Which courses do you want? ").split()

accepted, rejected = process_course_requests(course_catalog, student_requests)

print("Accepted: ")
for c in accepted:
  print(c)

print("Rejected:")
for c in rejected:
  print(c)



_='''
After every step we ask the required courses and check the overlaps. in this exercise
we show that it changes according to order. and the student could only take CS101 and PHY101.
'''

_='''EXERCISE 1
everytime we check latest timelines so which one we write first the other will not be counted.

What are the codes of the courses? CS101 MATH200 PHYS101 ART101 BIO101 CHEM101
What are the start times of the courses? (e.g 900) 900 1000 1300 1030 900 1000
What are the end times of the courses? (e.g 1100) 1100 1200 1500 1130 1200 1100
Which courses do you want? BIO101 CHEM101
Accepted:
BIO101

What are the codes of the courses? CS101 MATH200 PHYS101 ART101 BIO101 CHEM101
What are the start times of the courses? (e.g 900) 900 1000 1300 1030 900 1000
What are the end times of the courses? (e.g 1100) 1100 1200 1500 1130 1200 1100
Which courses do you want? CHEM101 BIO101
Accepted:
CHEM101

EXERCISE 2
#for checking exercise 2 we must consider all the courses in the time periods. even though
#starting and ending times are different, if one of the course's time period is in other course's
#timeline then there is a conflict. to prevent this we must add the expression end_new <= start_exist.

What are the codes of the courses? LIT200 HIST300
What are the start times of the courses? (e.g 900) 1300 1400
What are the end times of the courses? (e.g 1100) 1600 1500
Which courses do you want? LIT200 HIST300
Accepted:
LIT200


EXERCISE 3
in this case if when a course finishes then the other course starts at the same time
then it should not be conflict. for instance 08.00-09.30 course and 09.30-11.00 course
do not conflict because they do not share the same time period.

What are the codes of the courses? MATH101 PHYS101
What are the start times of the courses? (e.g 900) 800 930
What are the end times of the courses? (e.g 1100) 930 1100
Which courses do you want? MATH101 PHYS101
Accepted:
MATH101
PHYS101

EXERCISE 4
if a course conflict with multiple courses we stop the execution at the first conflict
by break code. the number of the conflicts do not matter in this situation.

What are the codes of the courses? CS101 MATH200 PHYS101 ART101
What are the start times of the courses? (e.g 900) 900 1000 1100 930
What are the end times of the courses? (e.g 1100) 1000 1100 1200 1130
Which courses do you want? CS101 MATH200 PHYS101 ART101
Accepted:
CS101
MATH200
PHYS101
Rejected:
ART101 rejected due to confliction with CS101.

'''


```
